```python
def load_ds(filepath, col1, col2, col3):
    ds = pd.read_csv(filepath, usecols=[col1, col2, col3])
    
    Vl = ds[col1]
    Vi = ds[col2]
    Vr = ds[col3]
    
    El = defaultdict(list)
    Er = defaultdict(list)
    for c1, c2, c3 in ds.values:
        El[c1].append(c2)
        Er[c3].append(c2)
    
    return Vl, Vi, Vr, dict(El), dict(Er)

print(load_ds('./Dataset/ds_salaries.csv', ))
```
